```python
def find_doc(this_issue):
    print("find_doc: ",this_issue)
    if this_issue.tracker.id == req_doc_tracker_id:
        print("retorno this", this_issue.subject)
        return this_issue.subject,this_issue.custom_fields.get(req_prefix_cf_id).value 

    # not do found yet
    current_parent = getattr(this_issue, 'parent', None)
    if current_parent is None:
        print("retorno none")
        return "",""
    
    else:
        parent_issue = redmine.issue.get(current_parent.id)
        print("Llamo al padre")
        return find_doc(parent_issue)
```

```python
idents = {}
for issue in my_doc_issues:
    idents[issue.id] = 1
    
print(idents)
    
for my_issue in my_project_issues:
    reqname = my_issue.subject
    thisdocident,thisprefix = find_doc(my_issue)
    doclist = redmine.issue.filter(project_id=req_project_id_str, subject=thisdocident, tracker_id=req_doc_tracker_id, status_id='*')
    if(len(doclist)>0):
        # Existe el documento asociado a la pestaña, puedo tratarlo
        thisdoc = doclist[0]
        newident = thisprefix+"{:04d}".format(idents[thisdoc.id])
        print("reqname: ",reqname," doc: ",thisdoc.subject," id_to_use: ", newident)
        idents[thisdoc.id] = idents[thisdoc.id] + 1
        #my_issue.subject = newident
        redmine.issue.update(resource_id=my_issue.id, subject = newident)
        
    else:
        print("ERROR")
    
```
